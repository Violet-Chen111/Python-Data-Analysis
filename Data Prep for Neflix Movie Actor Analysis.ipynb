# Data Preparation on the Scores for Netflix Movie Actors 

## Analysis objective

The objective of this analysis is to pre-process the data about actors' average IMDB scores on movies they participate in according to different genres, wuch as comendy, sci-fi, etc.

This project is to practice data pre-processing, so that the data is ready for analysis.

## Basic Info

The original dataset records all Netflix series and movies that is available within the US before July 2020. The dataset includes two data tables: `titles.csv`和`credits.csv`。

`titles.csv` includes the relevant information of each movie and series, including item ID, titles, genres, descriptions, show_type, etc. `credits.csv` includes information of over 70,000 directors and actors that have been appeared in Netflix series and movies, including names, item ID, character name, etc.

Meaning of each column in `titles.csv`:
- id：item ID
- title：item title
- show_type：show type, series/movie
- description：brief description
- release_year：released year
- age_certification：age certification
- runtime：the length of each episode the series or movie
- genres：genre list
- production_countries：production countries list
- seasons：if the item is series, total number of seasons
- imdb_id：item ID on IMDB
- imdb_score：IMDB score
- imdb_votes：IMDB votes
- tmdb_popularity：TMDB popularity
- tmdb_score：TMDB score

Meaning of each column in `credits.csv`:
- person_ID：staff ID
- id：ID of item they take part in
- name：name
- character_name：character name
- role：staff type, actor/director

## Organize the Data

Data organization closely connect to the analysis direction. The goal of ours is to dig out the actors from highly scored movies or series in different genres.

In order to gain the information of genres and actors simultaneously, we need to concatenate `cleaned_titles` and `cleaned_credits` according to `id`. That is because the `id` in both dataframe means the item ID.

```python
credits_with_titles = pd.merge(cleaned_credits, cleaned_titles, on="id", how="inner")
```

After we concatenate the two dataframes, we can gain the information of workpiece each actor participate in.

credits_with_titles.head()

Since we only consider the performance of actors, not the directors, we filter the observation with `ACTOR` in `role`.

```python
actor_with_titles = credits_with_titles.query('role == "ACTOR"')
```

In order to get actors with a high IMDB score in different genres, we need to group the data with genres and actors.

While we group the actors, we use `person_id` instead of `name` because different actors can possibly have identical names.

```python
groupby_genres_and_person_id = actor_with_titles.groupby(["genres", "person_id"])
```

After grouping, we will implement some aggregation calculation toward `imdb_score`, hence we only need to extract variable `imdb_score`. Then, we use `mean` to valculate the average IMDB scores of each actor.

```python
imdb_score_groupby_genres_and_person_id = groupby_genres_and_person_id["imdb_score"].mean()
imdb_score_groupby_genres_and_person_id
```

Then we use `reset_index` to reset the layered index, in order to get a more readible data frame.

```python
imdb_score_groupby_genres_and_person_id_df = imdb_score_groupby_genres_and_person_id.reset_index()
imdb_score_groupby_genres_and_person_id_df
```

Now we have done the data organization process for IMDB score data on genre and actor.

Nevertheless, we can still do some further pre-processing, such as a more detailed grouping on the output above. Then, we can find out the highest score of actors' work in different genre and the person that gets this score.

To get this result, we need to group the data according to `genres` again, and then extract the variable `imdb_score` to get its maximum value.

```python
genres_max_scores = imdb_score_groupby_genres_and_person_id_df.groupby("genres")["imdb_score"].max()
genres_max_scores
```

After we get the highest score, we concatenate this result with `imdb_score_groupby_genres_and_person_id_df` again. This action will give us the person ID of the highest score.

```python
genres_max_score_with_person_id = pd.merge(imdb_score_groupby_genres_and_person_id_df, genres_max_scores, on=["genres", "imdb_score"])
genres_max_score_with_person_id
```

From the result above, the highest score can gained by more than one actor.

To get the name of the person ID, we can connect this output with `cleaned_credits`. As we only need `person_id` and `name`, we can extract these two columns first and delete the duplicate rows.

```python
actor_id_with_names = cleaned_credits[['person_id', 'name']].drop_duplicates()
actor_id_with_names.head(10)
```

The next step is to connect `actor_id_with_names` with `genres_max_score_with_person_id`. We add the variable `name` to display the name of the actor who gets the highest average score.

```python
genres_max_score_with_actor_name = pd.merge(genres_max_score_with_person_id, actor_id_with_names, on="person_id")
genres_max_score_with_actor_name
```

In order to place the same genre together, we use `sort_values` method to sort the output according to `genres`. Then, we reset the index with `reset_index`.

After we reset the index, there will be an extra column `index`, which we can delete later.

```python
genres_max_score_with_actor_name = genres_max_score_with_actor_name.sort_values("genres").reset_index().drop("index", axis=1)
genres_max_score_with_actor_name
```
