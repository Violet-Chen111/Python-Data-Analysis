# Evaluate and Clean Sales Data from a E-commerce Company in the UK

## 1. Analysis objectives

The objective of this analysis project is to dig out the popular product according to the sales data. Therefore, the company can create a decent marketing strategy to improve sales revenue.

## 2. Basic Info

This dataset notes down every order detail of a online retail company in the UK from December 1st 2010 to December 9th 2011, covering the company's business data in different countries. This company mainly sells gifts that satisfying numerous context, including birthday gifts, anniversary gifts, Chirstmas gifts, etc. The main clientele of this company includes wholesalers, which takes up bigger proportion, and individual buyers.

The meanings of each column:
- `InvoiceNo`: Invoice number, a 6-digit number as the unique label of each order. If this code starts with 'C', it means this order is canceled.
- `StockCode`: Product code, a 5-digit number, as a unique label of each product.
- `Description`: The name of product.
- `InvoiceDate`: The issue date of invoice, stands for the ordering date and time.
- `UnitPrice`: Unit price, currency is pound(£).
- `CustomerID`: Customer ID, a 5-digit number, as a unique label of each customer.
- `Country`: Country name, the residence country of each customer.

## 3. Read the data

```python
import pandas as pd

original_data = pd.read_csv('e_commerce.csv')
original_data.head()
```

## 4. Evaluation

```python
original_data.sample(10)
```

### 4.1 Evaluate if the data is clean

```python
original_data.info()
```

##### According to the output, there are 541909 observation values. There are null values under `Description` and `CustomerID`.
##### Moreover, the data type of `InvoiceDate` should be date and time, while `CustomerID` should be strings. Data type transformation should be processed later.

### 4.2 Evaluate the missing values

```python
original_data[original_data['Description'].isnull()]
```

##### There are 1454 missing values in `Description`.
##### From the output, the values that miss `Description` value show a 0 in `UnitPrice`. For testing purpose, an extra filtering condition is added in order to confirm if all rows missing `Description` values has a '0' in `UnitPrice`.

```python
original_data[(original_data['Description'].isnull()) & (original_data['UnitPrice'] != 0)]
```

##### The output shows 0, meaning that the rows missing `Description` values do not contain valid `UnitPrice` values.
##### `Description` and `UnitPrice` are two variables that will contribute heavily to the following analysis process. If they are both missing/invalid, the data piece no longer perform its validity, hence it should be deleted later.

##### `CustomerID`contains null values too, hence we extract the null value first.

```python
original_data[original_data['CustomerID'].isnull()]
```

##### `CustomerID` means customer labels, which is not a crucial variable in the later analysis process. Besides, according to the output, some data pieces that miss `CustomerID` value are still valid, hence we keep those observation values. 

### 4.3 Evaluate the duplicate values

##### According to the meaning of each variable, `InvoiceNo`, `StockCode` and `CustomerID` are unique labels. However, a single order may contain numerous product, hence `InvoiceNo` is allowed to have duplicate values; different orders may have same product, hence `StockCode` is allowed to have duplicate values; a single customer may place more than one order and purchase more than one product, hence `CustomerID` is allowed to have duplicate value.
##### Then, for this dataset, we do not need to evaluate the duplicate value.

### 4.4 Evaluate inconsistant data

##### Inconsistancy is possible in `Country` variable, we need to check if different values means the same country.

```python
original_data['Country'].value_counts()
```

##### From the variable `Country`, 'USA' and 'United States' both stands for America, while 'United Kingdom', 'UK' and 'U.K.' all stands for United Kingdom. Hence, we need to keep only one value for each country.

### 4.5 Evaluate invalid and wrong data

```python
original_data.describe()
```

##### From the results, the variable 'Quantity' and 'UnitPrice' contain negative values, which will impact the upcoming analysis process. Hence, we will extract the negative values in 'Quantity' and interpret their meanings.

```python
original_data[original_data['Quantity'] < 0 ]
```

##### From the extract result, we can see the negative value in 'Quantity' comes with a 'InvoiceNo' that starts with 'C', which means a canceled order. 
##### But this inference is not necessarily correct. For the purpose of testing, we will add a filter condition to see if there is some negative quantities that do not show an invoice number starts with 'C.'

```python
original_data[(original_data['Quantity'] < 0) & (original_data['InvoiceNo'].str[0] != 'C') ]
```

##### The inspection above is proved correct. 
##### However, from the results above, the filtered values come with `UnitPrice` as 0. Therefore, a condition related to `UnitPrice` is added to test.

```python
original_data[(original_data['Quantity'] < 0) & (original_data['InvoiceNo'].str[0] != 'C') & (original_data['UnitPrice'] != 0) ]
```

##### According to the output, the inspection got proved correct. When `Quantity` is negarive, the value satisfy one of the following conditions:
- `InvoiceNo` starts with 'C', which stands for a canceled order.
- `UnitPrice` is 0, meaning that the price is 0 pound.
##### Those data points are not valid, hence no sales contribution. They should not be included in the following analysis process. So, we will delete the rows with negative `Quantity` values.

##### Next, we will extract the rows with a negative `UnitPrice`.

```python
original_data[original_data['UnitPrice'] < 0]
```

##### From the output above, the rows with negative`UnitPrice` shows a description of /Adjust bad debt/, not a actual trading order. Therefore, these could be deleted before the analysis process.

## 5. Data Cleaning

##### From the results of the previous evaluation part, the data cleaning includes the following steps:
- Transfer the data type of `InvoiceDate` to date and time.
- Transfer the data type of `CustomerID` to strings.
- Delete the null value in `Description`.
- Change the 'USA' in `Country` column to 'United States'.
- Change the 'UK' and 'U.K.' in `Country` to 'United Kingdom'.
- Delete the negative values in `Quantity`.
- Delete the negative values in `UnitPrice`.

##### To distinguish the dataset after cleaning, we create a new dataframe as 'cleaned_data', making it a copy of 'original_data'. The following cleaning process will be done in 'cleaned_data'.

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

### 5.1 Transfer the data type of `InvoiceDate` to date and time

```python
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data['InvoiceDate'])
cleaned_data['InvoiceDate']
```

### 5.2 Transfer the data type of `CustomerID` to strings

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].astype(str)
cleaned_data['CustomerID']                                                               
```

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].str.slice(0, -2)
cleaned_data['CustomerID'] 
```

### 5.3 Delete the null value in `Description`

```python
cleaned_data.dropna(subset = ['Description'], inplace = True)
```

```python
cleaned_data['Description'].isnull().sum()
```

### 5.4 Change the 'USA' in `Country` column to 'United States'

```python
cleaned_data['Country'] = cleaned_data['Country'].replace({'USA': 'United States'})
```

```python
len(cleaned_data[cleaned_data['Country'] == 'USA'])
```

### 5.5 Change the 'UK' and 'U.K.' in `Country` to 'United Kingdom'

```python
cleaned_data['Country'] = cleaned_data['Country'].replace({'UK' : 'United Kingdom', 'U.K.' : 'United Kingdom'})
```

```python
print(len(cleaned_data[cleaned_data['Country'] == 'UK']))
print(len(cleaned_data[cleaned_data['Country'] == 'U.K.']))
```

### 5.6 Delete the negative values in `Quantity`

```python
cleaned_data = cleaned_data[cleaned_data['Quantity'] >= 0]
```

```python
len(cleaned_data[cleaned_data['Quantity'] < 0])
```

### 5.7 Delete the negative values in `UnitPrice`

```python
cleaned_data = cleaned_data[cleaned_data['UnitPrice'] >= 0]
```

```python
len(cleaned_data[cleaned_data['UnitPrice'] < 0])
```

## 6. Save the dataset after cleaning

```python
cleaned_data.to_csv('e_commerce_cleaned.csv', index = False)
```
